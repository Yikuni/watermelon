```python
import torch
from torch import nn
from matplotlib import pyplot as plt
```

## 生成数据集
生成50个符合平均分布的x，x为1维数据，y和x满足下面的映射关系:
$$
y_{i} = 2\sin (x_{i}) + x_{i} ^{0.8} + \epsilon_{i}
$$
其中$\epsilon$ 是随机误差

```python
n_train = 50
x_train, _ = torch.sort(torch.rand(n_train)*5)

def f(x):
    return 2 * torch.sin(x) + x**0.8

y_train = f(x_train) + torch.normal(0.0, 0.5, (n_train,))
x_test = torch.arange(0, 5, 0.1)
y_true = f(x_test)
n_test = len(x_test)
```

```python
plt.plot(x_train, y_train)
```

```python
# 绘制结果的函数
def plot_result(y_hat):
    plt.plot(x_test, y_true, color='g')
    plt.plot(x_test, y_hat, color='r')
```

```python
plot_result(y_train)
```

## 平均汇聚
$$ 
f(x) = \frac{1}{n} \sum_{i=1}^{n}y_{i}
$$

```python
y_hat = torch.repeat_interleave(y_train.mean(), n_test)
plot_result(y_hat)
```

## 非参数注意力汇聚
$$
f(x) = \sum_{i=1}^{n}\alpha (x,x{i})y_{i}
$$
其中 $\alpha$被称为注意力权重(attention weight)，f(x)也可以写成下面的形式
$$
f(x) = \sum_{i=1}^{n}\frac{K(x,x{i})}{\sum_{j=1}^{n}K(x,x_{i})}y_{i}
$$
其中K是核，该公式所描述的估计器被称为Nadaraya-Watson核回归，如果引入高斯核，那么该公式将变成下面的形式(推导步骤见课本)
$$
f(x) = \sum_{i=1}^{n}softmax(-\frac{1}{2}(x-x_{i})^2)y_{i}
$$

```python
# 重复是为了方便下面的矩阵运算
X_repeat = x_test.repeat_interleave(n_train).reshape((-1, n_train))
X_repeat
```

```python
attention_weights = nn.functional.softmax(-(X_repeat - x_train)**2 / 2, dim = 1)
y_hat = torch.matmul(attention_weights, y_train)
plot_result(y_hat)
```

```python
plt.imshow(attention_weights, cmap='hot', interpolation='nearest')
plt.colorbar()
```

## 带参数注意力汇聚
$$
f(x)=\sum_{i=1}^{n}softmax(-\frac{1}{2}((x-x_{i})\omega)^2)y{i}
$$
公式中的$\omega$是一个1维的数
- Q: $x$
- K: $x_{i}$
- V: $y_{i}$

```python
weights = torch.ones((2, 10)) * 0.1
unsqueezed_weights = weights.unsqueeze(1) # 在外面加一层
weights, unsqueezed_weights, unsqueezed_weights.shape
```

```python
unsqueezed_weights.squeeze() # 可以减少一层
```

```python
class NWKernelRegression(nn.Module):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.w = nn.Parameter(torch.rand((1,), requires_grad=True))

    def forward(self, queries, keys, values):
        # queries和attention_weights的形状为(查询个数，K-V pair个数), reshape的目的是为了能够一次forward多个Q
        queries = queries.repeat_interleave(keys.shape[1]).reshape((-1, keys.shape[1]))
        self.attention_weights = nn.functional.softmax(-((queries - keys) * self.w) ** 2 / 2, dim=1)
        return torch.bmm(self.attention_weights.unsqueeze(1), values.unsqueeze(-1)).reshape(-1)
```

```python
net = NWKernelRegression()
loss = nn.MSELoss(reduction='none')
trainer = torch.optim.SGD(net.parameters(), lr=0.5)
```

```python
loss_history = []
# X_tile的形状:(n_train，n_train)，每一行都包含着相同的训练输入
X_tile = x_train.repeat((n_train, 1))
# Y_tile的形状:(n_train，n_train)，每一行都包含着相同的训练输出
Y_tile = y_train.repeat((n_train, 1))
# keys的形状:('n_train'，'n_train'-1)
keys = X_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape((n_train, -1))
# values的形状:('n_train'，'n_train'-1)
values = Y_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape((n_train, -1))
for epoch in range(5):
    trainer.zero_grad()
    l = loss(net(x_train, keys, values), y_train)
    l.sum().backward()
    trainer.step()
    loss_history.append(l.sum().detach().numpy())
    print(f'epoch {epoch + 1}, loss {float(l.sum()):.6f}')
```

```python
import numpy as np
plt.plot(np.arange(1, len(loss_history)+1), loss_history)
```

```python
keys = x_train.repeat((n_test, 1))
values = y_train.repeat((n_test, 1))
y_hat = net(x_test, keys, values).unsqueeze(1).detach()
plot_result(y_hat)
```
