```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
```

```python
# torch.set_default_dtype(torch.float32)
if torch.cuda.is_available():
    device = torch.device("cuda:0")  # you can continue going on here, like cuda:1 cuda:2....etc. 
    print("Running on the GPU")
    torch.set_default_device(device)
```

```python
import iris_pytorch
training_data = iris_pytorch.IrisDataset()
```

> 这是自定义的数据集

```python
train_dataloader = torch.utils.data.DataLoader(training_data, batch_size=30, shuffle=True, generator=torch.Generator(device))
```

```python
for x, y in train_dataloader:
    print(x, y)
    print(x.dtype)
    break;
```

> Q: 如何以float32形式读取数据?
> A: 自定义数据集getItem函数返回np.array dtype=np.float32

```python
model = nn.Sequential(
    nn.Linear(4, 12),
    nn.ReLU(),
    nn.Linear(12, 3)
)
# def init_weights 这里省略, 因为默认会init的
```

```python
lr = 0.01
max_epochs = 30
```

```python
loss = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(params=model.parameters(), lr=lr, weight_decay=0.02, betas=(0.9, 0.99))
```

```python
history = []
for epoch in range(max_epochs):
    l = None
    for x, y in train_dataloader:
        l = loss(model(x), y.long())
        optimizer.zero_grad()
        l.backward()
        optimizer.step()
        print(f"\r epoch {epoch} loss = {l}", end='')
    history.append(l.cpu().detach().numpy())
    print("")
```

```python
plt.plot(np.arange(len(history)), history)
plt.show()
```

```python
from sklearn.datasets import load_iris
data = load_iris().data
result = torch.argmax(model(torch.tensor(data, dtype=torch.float32)).cpu(), 1).numpy()
```

```python
colors = np.where(result == 0, 'r', 'g')
colors = np.where(result == 1, 'g', colors)
colors = np.where(result == 2, 'b', colors)
data = np.array(data)
plt.scatter(data[:, 0], data[:, 1], color=colors)
plt.show()
```

```python
acc = np.where(result == load_iris().target, 1, 0).sum() / 150
acc
```
